## HTML소스코드 -> 파이썬 객체로 변환

```python
keyword = input('검색할 키워드 입력: ')

url = 'https://search.naver.com/search.naver?'

res = req.get(url, params={'query':keyword})

# 파싱: 웹 페이지 내 사용자가 원하는 데이터를 분석하여 추출
# 파서 : lxml, html.parser, html5lib (속도순)
# bs(변환할 데이터, 파서)
html = bs(res.text, 'html.parser')

print(html)
print(type(html))
```

```python
# 나무위키 접근
# 크롬에서 해당 하이퍼링크 우클릭 검사 -> ctrl + f -> (클래스).link_tit -> 첫번째
html.select_one('.link_tit') # css 선택자
```

```python
# 요소 내용 접근
html.select_one('.link_tit').text.strip() # 공백제거 strip
```

```python
# 여러 개의 요소에 접근
# select : 리스트로 반환
html.select('.api_txt_lines.total_tit')
```

```python
for view_tit in html.select('.api_txt_lines.total_tit') :
    print(view_tit.text)
```

## 멜론 일간 차트 1-100위 정보 수집
- https://www.melon.com/chart/day/index.htm
- 곡명, 가수, 앨범

```python
url = 'https://www.melon.com/chart/day/index.htm'

header = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.141 Safari/537.36',
    'cookie':'NNB=UKQWICXJMXQV6; nx_ssl=2; NRTK=ag#20s_gr#2_ma#-2_si#0_en#2_sp#2; page_uid=hsHC7dp0JXossPTppKhssssss8h-316163'
}
    
res = req.get(url, headers = header)
# 2. bs 라이브러리로 응답받은 HTML 데이터 -> 파이썬 객체로 변환
html = bs(res.text, 'html.parser')
html
```

```python
songnames = html.select("div.ellipsis.rank01")
songnames
```

```python
singers = html.select("div.ellipsis.rank02 > a")

singers
```

```python
albums = html.select("div.ellipsis.rank03")
albums.strip()
```

```python
song_list = []
sing_list = []
al_list = []
for i in range(len(songnames)) :
    print('{}. 노래제목: {} 가수: {} 앨범: {}' .format(i+1,songnames[i].text.strip(),singers[i].text.strip(),albums[i].text.strip()))
    song_list.append(songnames[i].text.strip())
    sing_list.append(singers[i].text.strip())
    al_list.append(albums[i].text.strip())
```

```python
import pandas as pd
df = pd.DataFrame({ 'Songname': song_list, \
    'Singer': sing_list, \
    'Album': al_list
})
```

```python
df.to_csv('melon_top100.csv', encoding="utf-8-sig")
```
