```python
import matplotlib.pyplot as plt
```

```python
y = [2,4,6]
# x = [0,1,2] 0부터 시작

# plot() : line chart 그리는 메소드
plt.plot(y)
plt.show()
```

```python
y = [2,4,6]
x = [3,4,5]  # 3부터 시작

# plot() : line chart 그리는 메소드
plt.plot(x,y)
plt.show()
```

```python
y = [2,4,6]
x = [3,6,2]  # 3부터 시작

# plot() : line chart 그리는 메소드
plt.plot(x,y)
plt.show()
```

```python
y = [2,4,6]
# x = [0,1,2] 0부터 시작

# plot() : line chart 그리는 메소드
plt.plot(y, ls='--')
plt.show()
```

```python
y = [2,4,6]
# x = [0,1,2] 0부터 시작

# plot() : line chart 그리는 메소드
plt.plot(y, marker = 'o')
plt.show()
```

```python
y = [2,4,6]
# x = [0,1,2] 0부터 시작

# plot() : line chart 그리는 메소드
plt.plot(y, color = 'r')
plt.show()
```

```python
# 이상치(outlier) 없애고 그리고 싶을때
plt.plot(y)
plt.xlim(-10,20) # 범위
plt.ylim(0, 10)
plt.show()
```

```python
plt.plot(y)
plt.xticks([0,0.5,4]) # 아래 표시되는 숫자값
plt.yticks([0,2,6])
plt.grid() # tick에 맞춰 선을 그어줌
plt.show()
```

# 여러개 플롯

```python
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]
plt.plot(x)
plt.plot(y)
plt.plot(z)
plt.show()
```

```python
plt.plot(x)
plt.plot(y)
plt.show()

plt.plot(z)
plt.show()
```

```python
plt.plot(x, label = 'x')
plt.plot(y, label = 'y')
plt.plot(z, label = 'z')

plt.legend()
plt.show()
```
