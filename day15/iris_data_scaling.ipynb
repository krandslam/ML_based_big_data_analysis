```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
```

```python
iris = load_iris()
iris
```

```python
iris_data = iris['data']
```

```python
iris_df = pd.DataFrame(data = iris_data,
                       columns = iris['feature_names']
)
iris_df.head()
```

# Standard scaler
- 평균 0, 분산 1

```python
from sklearn.preprocessing import StandardScaler

# import 한 StandardScaler를 불러와서 변수에 담아줌
scaler = StandardScaler()
# iris 데이터 프레임으로 스케일러를 학습
scaler.fit(iris_df)
# 학습된 스케일러의 데이터값을 실제로 변환시켜줌
iris_scaler_SS = scaler.transform(iris_df)
```

```python
# numpy의 array 형태로 변환 되어있음
iris_scaler_SS 
```

```python
iris_df_scaler_SS = pd.DataFrame(data = iris_scaler_SS, columns = iris['feature_names'])
iris_df_scaler_SS.head()
```

```python
iris_df.mean()
```

```python
iris_df_scaler_SS.mean()
```

# MinMaxScaler
- 0 ~ 1 사이의 값으로 변환 (-가 있다면 -1~ 1 사이 값으로 변환)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler1 = MinMaxScaler()
scaler1.fit(iris_df)
iris_scaler_MM = scaler1.transform(iris_df)
iris_scaler_MM
```

```python
iris_df_scaler_MM = pd.DataFrame(data = iris_scaler_MM, columns = iris['feature_names'])
iris_df_scaler_MM.head()
```

```python
print(iris_df_scaler_MM.min())
print(iris_df_scaler_MM.max())
```
